```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
data = pd.read_csv("/home/work/.code/CoAtNet/pytorch-image-models/output/train/14626_no_resize_50epoch_cosinerestartsLR/summary.csv")

# Epoch 값 가져오기
epochs = data['epoch']

# Train loss 및 eval loss 가져오기
train_loss = data['train_loss']
eval_loss = data['eval_loss']

# Eval top1 가져오기
eval_top1 = data['eval_top1']

# Learning rate 가져오기
learning_rate = data['lr']


# Train loss 및 eval loss 그래프
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_loss, label='Train Loss', marker='o')
plt.plot(epochs, eval_loss, label='Eval Loss', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Eval Loss')
plt.legend()
plt.grid(True)
plt.show()

# Eval top1 그래프
plt.figure(figsize=(10, 5))
plt.plot(epochs, eval_top1, label='Eval Top1', marker='o', color='orange')
plt.xlabel('Epoch')
plt.ylabel('Top1 Accuracy')
plt.title('Eval Top1 Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Learning Rate 그래프
plt.figure(figsize=(10, 5))
plt.plot(epochs, learning_rate, label='Learning Rate', marker='o', color='green')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Epoch Learning Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# file number checking

import os
from pathlib import Path

# 이미지 파일의 확장자를 리스트로 정의합니다.
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.gif', '.tiff']

# 이미지를 저장할 디렉토리
directory = Path('/usr/src/data/20240319_tsinghua_cleaned/test/others')

# 이미지 파일 수를 세는 변수
image_count = 0

# 디렉토리 내 파일을 탐색합니다.
for file in directory.iterdir():
    # 파일의 확장자가 이미지 확장자 목록에 포함되어 있는지 확인합니다.
    if file.suffix.lower() in image_extensions:
        image_count += 1

# 결과 출력
print(f"폴더 '{directory}' 안에 이미지 파일의 개수: {image_count}")
```
